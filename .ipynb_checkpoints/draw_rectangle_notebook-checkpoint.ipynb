```python
%matplotlib widget
import json
import time
import datetime
from dateutil.tz import tzlocal
import ipywidgets as widgets
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from IPython.display import display
from ati_eye_utils import im_show

from picamreader import PiCamReader
```

```python
image_width=640
image_height=480

cam_capture=PiCamReader(output_size=(image_width,image_height))
cam_capture.start()
```

```python
# capture an image
if cam_capture.is_new_data:
    pi_image, timestamp,frame_id=cam_capture.get_data()

# capture and display the image
im_show(pi_image)
```

```python
class InteractiveRectangle:
    def __init__(self, image):
        self.fig, self.ax = plt.subplots()
        self.image = image
        self.ax.imshow(self.image)
        self.rectangle = Rectangle((0,0), 1, 1, edgecolor='red', facecolor='none')
        self.ax.add_patch(self.rectangle)
        
        self.dragging = False
        self.x0, self.y0, self.width, self.height = 0, 0, 0, 0
        
        # Connect the mouse events to the corresponding handlers
        self.cid_press = self.fig.canvas.mpl_connect('button_press_event', self.on_press)
        self.cid_release = self.fig.canvas.mpl_connect('button_release_event', self.on_release)
        self.cid_motion = self.fig.canvas.mpl_connect('motion_notify_event', self.on_motion)

        # Confirmation button
        self.button = widgets.Button(description="Confirm Rectangle")
        self.button.on_click(self.save_rectangle_properties)
        display(self.button)
        
        self.output = widgets.Output()
        display(self.output)
        
    def on_press(self, event):
        if event.inaxes != self.ax:
            return
        self.dragging = True
        self.x0, self.y0 = event.xdata, event.ydata
        self.rectangle.set_width(0)
        self.rectangle.set_height(0)
        self.rectangle.set_xy((self.x0, self.y0))
        self.fig.canvas.draw()
        
    def on_motion(self, event):
        if not self.dragging:
            return
        x1, y1 = event.xdata, event.ydata
        self.width = abs(x1 - self.x0)
        self.height = abs(y1 - self.y0)
        self.rectangle.set_width(self.width)
        self.rectangle.set_height(self.height)
        self.fig.canvas.draw()
    
    def on_release(self, event):
        self.dragging = False
        x1, y1 = event.xdata, event.ydata
        self.width = abs(x1 - self.x0)
        self.height = abs(y1 - self.y0)
        self.rectangle.set_width(self.width)
        self.rectangle.set_height(self.height)
        self.fig.canvas.draw()
        
        self.output.clear_output()
        with self.output:
            if self.width == 0 or self.height == 0:
                print("Invalid rectangle: Please drag to draw the rectangle, do not just click. Try again.")
            else:
                print(f'Drawn Rectangle: (x0={self.x0:.1f}, y0={self.y0:.1f}, width={self.width:.1f}, height={self.height:.1f})')
    
    def save_rectangle_properties(self, _):
        if self.width == 0 or self.height == 0:
            with self.output:
                print("Cannot save: Rectangle dimensions are zero. Please draw a valid rectangle.")
            return
        current_time = datetime.datetime.now(tzlocal())
        time_str = current_time.strftime('%Y-%m-%dT%H:%M:%S%z')
        timezone_name = current_time.astimezone().tzname()
        properties = {
            'x0': self.x0,
            'y0': self.y0,
            'width': self.width,
            'height': self.height,
            'timestamp': time_str,
            'time_zone': timezone_name
        }
        with open('rectangle_properties.json', 'w') as file:
            json.dump(properties, file)
        with self.output:
            print("Rectangle properties saved!")

# Example usage
# path = "/Users/saikarthik/Downloads/lifter_01_frame_231.jpg"  # Change this to the path of your JPG file
# path = "test_image.jpg"  # Change this to the path of your JPG file
rect_tool = InteractiveRectangle(pi_image)
```
